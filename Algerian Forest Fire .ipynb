# Algerian Forest Fires Dataset
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attribues and 1 output attribue (class) The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observations
Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
RH : Relative Humidity in %: 21 to 90
Ws :Wind speed in km/h: 6 to 29
Rain: total day in mm: 0 to 16.8 FWI Components
Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
Drought Code (DC) index from the FWI system: 7 to 220.4
Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
Buildup Index (BUI) index from the FWI system: 1.1 to 68
Fire Weather Index (FWI) Index: 0 to 31.1
Classes: two classes, namely Fire and not Fire

# model training

```python
data1.head()
```

```python
#independent and dependent features
x=data1.drop("FWI",axis=1)
y=data1["FWI"]
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
x_train.shape,x_test.shape
```

```python
plt.figure(figsize=(20,8))
corr=x_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,threshold):
    col_corr=set()                            #if the independent feature is more co-related than a threshold then we drop the feature
    col_matrix=dataset.corr()                 #this solves the multi colinearity problem
    for i in range(len(col_matrix.columns)):
        for j in range(i):
            if abs(col_matrix.iloc[i,j])>threshold:
                col_name=col_matrix.columns[i]
                col_corr.add(col_name)
    return col_corr
```

```python
corr_features=correlation(x_train,0.85)
```

```python
x_train.drop(corr_features,axis=1,inplace=True)
x_test.drop(corr_features,axis=1,inplace=True)
```

```python
x_train.shape,x_test.shape
```

# Feature scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
#box plot to understand effect of standard scalar
sns.boxplot(data=x_train)
plt.title("x_train before scaling")
```

```python
sns.boxplot(data=x_train_scaled)
plt.title("x_train after scaling")
```

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
linreg=LinearRegression()
linreg.fit(x_train_scaled,y_train)
y_pred=linreg.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print("r2_score ",score)
```

```python
plt.scatter(y_test,y_pred) #if the points are linear then it is a good model
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
lasso=Lasso()
lasso.fit(x_train_scaled,y_train)
y_pred=lasso.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
print("the mean absolute error is",mae)
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print("r2_score ",score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
ridge=Ridge()
ridge.fit(x_train_scaled,y_train)
y_pred=ridge.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
print("the mean absolute error is",mae)
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print("r2_score ",score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
en=ElasticNet()
en.fit(x_train_scaled,y_train)
y_pred=en.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
print("the mean absolute error is",mae)
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print("r2_score ",score)
plt.scatter(y_test,y_pred)
```
